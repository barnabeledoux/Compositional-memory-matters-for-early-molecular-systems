```python
class disttodata:
    def __init__(self, data, timedata, theo, timetheo):
        self.data, self.tdata, self.theo, self.ttheo = data, timedata, theo, timetheo
        indcom, i, tdat = [], 0, timedata[0]
        for j, tth in enumerate(timetheo):
            if tth>=tdat and i<len(timedata):
                i+=1
                if i<len(timedata):
                    tdat = timedata[i]
                indcom.append(j)
        self.indcomp = indcom
        self.l = min(len(indcom), len(timedata))
    
    def dist(self):
        norm = 0
        for tind in range(self.l):
            id, it = tind, self.indcomp[tind]
            norm += (self.data[id]-self.theo[it])**2/(((self.data[id]+self.theo[it])/2)**2)
        return np.sqrt(norm/self.l)

def smoothen(list, nm = 10):
        reslist = []
        meanlist = [k-nm//2 for k in range(nm)]
        for j in range(len(list)):
            reslist.append((sum([list[min(max(0,j+k),len(list)-1)] for k in meanlist]))/nm)
        return reslist
```

```python
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

cmap_name = 'phasecolors'
n_bins = []

alpha = np.array([[5.E-1,0.],[0.,5.E-1]])
gamma = np.array([[1.E1,5.E-1],[0.,5.E-1]])
mu=np.array([7.E-3,0.])
nu=np.array([7.E-3,7.E-3])
K = 3.E1
r=1

cat = 'bethedge1'

x0list = np.load(cd + '/data/x0list_'+cat+str(r)+'.npy')
x1list = np.load(cd + '/data/x1list_'+cat+str(r)+'.npy')
z0list = np.load(cd + '/data/z0list_'+cat+str(r)+'.npy')
z1list = np.load(cd + '/data/z1list_'+cat+str(r)+'.npy')
darr = np.load(cd + '/data/darr_'+cat+str(r)+'.npy')
Tmatarr = np.load(cd + '/data/Tmatarr_'+cat+str(r)+'.npy')*alpha[0][0]

for listfrac in [x0list,x1list,z0list,z1list]:
    for i,xline in enumerate(listfrac):
        for j,x in enumerate(xline):
            if x!=x:
                listfrac[i,j]=listfrac[i+1,j] if listfrac[i+1,j]==listfrac[i+1,j] else listfrac[i,j-1]

def replace_outliers_with_neighbors(data, threshold, border = 3):

    """
    Replaces points in the data that differ too much from their neighbors 
    with the average of the neighboring points.
    """
    clean_data = data.copy()
    rows, cols = data.shape[0], data.shape[0]
    for i in range(border, rows - border):  # Avoid edges for simplicity
        for j in range(border, cols - border):
            if not np.isnan(data[i, j]):  # Ignore NaN entries
                if border == 3:
                    neighbors = [
                        data[i-3, j], data[i-2, j-1], data[i-1, j-2], data[i, j-3], data[i-2, j+1], data[i-1, j+2], data[i, j+3], data[i+1, j-2], data[i+2, j-1], data[i+3, j], data[i+1, j+2], data[i+2, j+1],
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 2:
                    neighbors = [
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 1:
                    neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                elif border == 0:
                    if i == 0 or i == rows - 1:
                        if j == 0 or j == cols - 1:
                            if i == 0 and j == 0:
                                neighbors = [
                                    data[i+1, j], data[i, j+1], data[i+1, j+1]
                                ]
                            elif i == 0 and j == cols - 1:
                                neighbors = [
                                    data[i+1, j], data[i, j-1], data[i+1, j-1]
                                ]
                            elif i == rows - 1 and j == 0:
                                neighbors = [
                                    data[i-1, j], data[i, j+1], data[i-1, j+1]
                                ]
                            else :
                                neighbors = [
                                    data[i-1, j], data[i, j-1], data[i-1, j-1]
                                ]
                        else:
                            neighbors = [
                                data[i, j-1], data[i, j+1], data[i + (i == 0) - (i == cols - 1), j]
                            ]
                    elif j == 0 or j == cols - 1:
                        neighbors = [
                            data[i-1, j], data[i+1, j], data[i, j + (j == 0) - (j == rows - 1)]
                        ]
                    else :
                        neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                neighbors = [x for x in neighbors if not np.isnan(x)]  # Exclude NaNs
                avg_neighbors = np.mean(neighbors) if neighbors else data[i, j]
                if abs(data[i, j] - avg_neighbors) > threshold:
                    clean_data[i, j] = avg_neighbors
                    #print(data[i, j], clean_data[i, j])
    return clean_data

# Define a threshold for determining outliers
threshold = 0.005 # Adjust based on acceptable difference

# Apply the cleaning to all data lists
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=3))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=2))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=1))

for i in range(6):
    x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
    data_list = []
    for data in [x0list, x1list, z0list, z1list]:
        #print(data)
        data_list.append(replace_outliers_with_neighbors(data, threshold, border=0))



X,Y = np.meshgrid(Tmatarr, darr)

ylim1 = 5.E0
ylim2 = 4.E1
xlim1 = 1.E0
xlim2 = 1.5E1

xtitle = r'$\alpha_{0,0} T_{mat}$'
ytitle = r'$d$'
title = r'$K_r=$'+r'{}'.format(K) + r'$, \,K_p=$'+r'{}'.format(r*K) + r' $, \,\mu = {}$ '.format(mu[0]) + r' $h^{-1}$,' + r' $\nu = {}$'.format(nu[0]) + r'  $h^{-1}$'

clist = (carr[:7]+ carr[8:-2]).copy() #carr[:13].copy()
clist.reverse()
clist = [(0, clist[1]), (0.03, clist[1]), (0.1, clist[2]), (0.2, clist[3]), (0.38, clist[4]), (0.5, clist[5]), (0.6, clist[6]), (0.7, clist[7]), (0.75, clist[8]), (0.82, clist[9]), (1., clist[10])] #[(0, clist[0]), (0.18, clist[1]), (0.25, clist[2]), (0.31, clist[3]), (0.38, clist[4]), (0.48, clist[5]), (0.5, clist[6]), (0.6, clist[7]), (0.65, clist[8]), (0.7, clist[9]), (0.75, clist[10]), (0.82, clist[11]), (1., clist[12])]
cmapphase = LinearSegmentedColormap.from_list(cmap_name, clist)
#norm = mpl.colors.Normalize(vmin=0., vmax=1.)
# Create a single figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid of subplots
#fig.suptitle(title, fontsize=16)  # Main title for the figure

# List of data and labels
data_list = [(data_list[0], r'$x_0$'), (data_list[1], r'$x_1$'), (data_list[2], r'$z_0$'), (data_list[3], r'$z_1$')]
axes = axs.ravel()  # Flatten the axes array for easy iteration

# Loop through each subplot and plot
for ax, (data, label) in zip(axes, data_list):
    c = ax.pcolormesh(X, Y, data, cmap=cmapphase, vmin=0., vmax=1., shading='gouraud')
    cbar = fig.colorbar(c, ax=ax, label=label)
    cbar.set_label(label, fontsize=20)
    cbar.ax.tick_params(labelsize=13)
    ax.tick_params(which = 'both', labelsize = 13)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(xlim1, xlim2)
    ax.set_ylim(ylim1, ylim2)
    ax.set_xlabel(xtitle, fontsize=20)
    ax.set_ylabel(ytitle, fontsize=20)
    #ax.set_title(label, fontsize=16)
    
fig.set_size_inches(10., 8.)

# Adjust layout to prevent overlapping
fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # Leave space for the title

plt.savefig('results/phasediag_strongdilu_Td_'+cat+str(r)+'.pdf')
```

```python
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

cmap_name = 'phasecolors'
n_bins = []

alpha = np.array([[5.E-1,0.],[0.,5.E-1]])
gamma = np.array([[1.E1,5.E-1],[0.,5.E-1]])
mu=np.array([5.E-3,0.])
nu=np.array([5.E-3,5.E-3])
K = 3.E1

cat = 'bethedge_stirring'

x0list = np.load(cd + '/data/x0list_'+cat+str(r)+'.npy')
x1list = np.load(cd + '/data/x1list_'+cat+str(r)+'.npy')
z0list = np.load(cd + '/data/z0list_'+cat+str(r)+'.npy')
z1list = np.load(cd + '/data/z1list_'+cat+str(r)+'.npy')
darr = np.load(cd + '/data/darr_'+cat+str(r)+'.npy')
Tmatarr = np.load(cd + '/data/Tmatarr_'+cat+str(r)+'.npy')*alpha[0][0]

for listfrac in [x0list,x1list,z0list,z1list]:
    for i,xline in enumerate(listfrac):
        for j,x in enumerate(xline):
            if x!=x:
                listfrac[i,j]=listfrac[i+1,j] if listfrac[i+1,j]==listfrac[i+1,j] else listfrac[i,j-1]

def replace_outliers_with_neighbors(data, threshold, border = 3):
    """
    Replaces points in the data that differ too much from their neighbors 
    with the average of the neighboring points.
    """
    clean_data = data.copy()
    rows, cols = data.shape[0], data.shape[0]
    for i in range(border, rows - border):  # Avoid edges for simplicity
        for j in range(border, cols - border):
            if not np.isnan(data[i, j]):  # Ignore NaN entries
                if border == 3:
                    neighbors = [
                        data[i-3, j], data[i-2, j-1], data[i-1, j-2], data[i, j-3], data[i-2, j+1], data[i-1, j+2], data[i, j+3], data[i+1, j-2], data[i+2, j-1], data[i+3, j], data[i+1, j+2], data[i+2, j+1],
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 2:
                    neighbors = [
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 1:
                    neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                elif border == 0:
                    if i == 0 or i == rows - 1:
                        if j == 0 or j == cols - 1:
                            if i == 0 and j == 0:
                                neighbors = [
                                    data[i+1, j], data[i, j+1], data[i+1, j+1]
                                ]
                            elif i == 0 and j == cols - 1:
                                neighbors = [
                                    data[i+1, j], data[i, j-1], data[i+1, j-1]
                                ]
                            elif i == rows - 1 and j == 0:
                                neighbors = [
                                    data[i-1, j], data[i, j+1], data[i-1, j+1]
                                ]
                            else :
                                neighbors = [
                                    data[i-1, j], data[i, j-1], data[i-1, j-1]
                                ]
                        else:
                            neighbors = [
                                data[i, j-1], data[i, j+1], data[i + (i == 0) - (i == cols - 1), j]
                            ]
                    elif j == 0 or j == cols - 1:
                        neighbors = [
                            data[i-1, j], data[i+1, j], data[i, j + (j == 0) - (j == rows - 1)]
                        ]
                    else :
                        neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                neighbors = [x for x in neighbors if not np.isnan(x)]  # Exclude NaNs
                avg_neighbors = np.mean(neighbors) if neighbors else data[i, j]
                if abs(data[i, j] - avg_neighbors) > threshold:
                    clean_data[i, j] = avg_neighbors
                    #print(data[i, j], clean_data[i, j])
    return clean_data

# Define a threshold for determining outliers
threshold = 0.005 # Adjust based on acceptable difference

# Apply the cleaning to all data lists
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=3))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=2))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=1))

for i in range(6):
    x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
    data_list = []
    for data in [x0list, x1list, z0list, z1list]:
        #print(data)
        data_list.append(replace_outliers_with_neighbors(data, threshold, border=0))



X,Y = np.meshgrid(Tmatarr, darr)

ylim1 = 5.E0
ylim2 = 4.E1
xlim1 = 1.E0
xlim2 = 1.5E1

xtitle = r'$\alpha_{0,0} T_{mat}$'
ytitle = r'$d$'
title = r'$K=$'+r'{}'.format(K) + r' $\mu = {}$ '.format(mu[0]) + r' $h^{-1}$,' + r' $\nu = {}$'.format(nu[0]) + r'  $h^{-1}$'

clist = carr[:13].copy()
clist.reverse()
clist = [(0, clist[0]), (0.05, clist[1]), (0.1, clist[2]), (0.2, clist[3]), (0.32, clist[4]), (0.4, clist[5]), (0.5, clist[6]), (0.6, clist[7]), (0.65, clist[8]), (0.7, clist[9]), (0.75, clist[10]), (0.82, clist[11]), (1., clist[12])] #[(i*1/11., clist[i]) for i in range(len(clist))]
cmapphase = LinearSegmentedColormap.from_list(cmap_name, clist)
#norm = mpl.colors.Normalize(vmin=0., vmax=1.)
# Create a single figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid of subplots
fig.suptitle(title, fontsize=16)  # Main title for the figure

# List of data and labels
data_list = [(data_list[0], r'$x_0$'), (data_list[1], r'$x_1$'), (data_list[2], r'$z_0$'), (data_list[3], r'$z_1$')]
axes = axs.ravel()  # Flatten the axes array for easy iteration

# Loop through each subplot and plot
for ax, (data, label) in zip(axes, data_list):
    c = ax.pcolormesh(X, Y, data, cmap=cmapphase, vmin=0., vmax=1., shading='gouraud')
    fig.colorbar(c, ax=ax, label=label)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(xlim1, xlim2)
    ax.set_ylim(ylim1, ylim2)
    ax.set_xlabel(xtitle, fontsize=12)
    ax.set_ylabel(ytitle, fontsize=12)
    ax.set_title(label, fontsize=14)
    
fig.set_size_inches(8., 8.)

# Adjust layout to prevent overlapping
fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # Leave space for the title

plt.savefig('results/phasediag_strongdilu_Td_'+cat+str(r)+'.pdf')
```

```python
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

cmap_name = 'phasecolors'
n_bins = []

K = 3.E1

ratioalp0alp1 = 1.
ratioalp0gam0 = 10.
ratioalp0gam01 = 5.
ratioalp0gam1 = 5.

x0list = np.load(cd + '/data/x0listmualp_gen'+str(r)+'.npy')
x1list = np.load(cd + '/data/x1listmualp_gen'+str(r)+'.npy')
z0list = np.load(cd + '/data/z0listmualp_gen'+str(r)+'.npy')
z1list = np.load(cd + '/data/z1listmualp_gen'+str(r)+'.npy')
muarr = np.load(cd + '/data/muarrmualp_gen'+str(r)+'.npy')
alparr = np.load(cd + '/data/alparrmualp_gen'+str(r)+'.npy')

for listfrac in [x0list,x1list,z0list,z1list]:
    for i,xline in enumerate(listfrac):
        for j,x in enumerate(xline):
            if x!=x:
                listfrac[i,j]=listfrac[i+1,j] if listfrac[i+1,j]==listfrac[i+1,j] else listfrac[i,j-1]

def replace_outliers_with_neighbors(data, threshold, border = 3):
    """
    Replaces points in the data that differ too much from their neighbors 
    with the average of the neighboring points.
    """
    clean_data = data.copy()
    rows, cols = data.shape[0], data.shape[0]
    for i in range(border, rows - border):  # Avoid edges for simplicity
        for j in range(border, cols - border):
            if not np.isnan(data[i, j]):  # Ignore NaN entries
                if border == 3:
                    neighbors = [
                        data[i-3, j], data[i-2, j-1], data[i-1, j-2], data[i, j-3], data[i-2, j+1], data[i-1, j+2], data[i, j+3], data[i+1, j-2], data[i+2, j-1], data[i+3, j], data[i+1, j+2], data[i+2, j+1],
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 2:
                    neighbors = [
                        data[i-2, j], data[i-1, j-1], data[i, j-2], data[i-1, j], data[i+1, j], data[i+2, j], data[i+1, j+1], data[i, j+2],
                        data[i, j-2], data[i-1, j+1], data[i, j-1], data[i, j+1], data[i+1, j-1], data[i, j+2]
                    ]
                elif border == 1:
                    neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                elif border == 0:
                    if i == 0 or i == rows - 1:
                        if j == 0 or j == cols - 1:
                            if i == 0 and j == 0:
                                neighbors = [
                                    data[i+1, j], data[i, j+1], data[i+1, j+1]
                                ]
                            elif i == 0 and j == cols - 1:
                                neighbors = [
                                    data[i+1, j], data[i, j-1], data[i+1, j-1]
                                ]
                            elif i == rows - 1 and j == 0:
                                neighbors = [
                                    data[i-1, j], data[i, j+1], data[i-1, j+1]
                                ]
                            else :
                                neighbors = [
                                    data[i-1, j], data[i, j-1], data[i-1, j-1]
                                ]
                        else:
                            neighbors = [
                                data[i, j-1], data[i, j+1], data[i + (i == 0) - (i == cols - 1), j]
                            ]
                    elif j == 0 or j == cols - 1:
                        neighbors = [
                            data[i-1, j], data[i+1, j], data[i, j + (j == 0) - (j == rows - 1)]
                        ]
                    else :
                        neighbors = [
                        data[i-1, j], data[i+1, j], data[i, j-1], data[i, j+1]
                    ]
                neighbors = [x for x in neighbors if not np.isnan(x)]  # Exclude NaNs
                avg_neighbors = np.mean(neighbors) if neighbors else data[i, j]
                if abs(data[i, j] - avg_neighbors) > threshold:
                    clean_data[i, j] = avg_neighbors
                    #print(data[i, j], clean_data[i, j])
    return clean_data

# Define a threshold for determining outliers
threshold = 0.005 # Adjust based on acceptable difference

# Apply the cleaning to all data lists
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=3))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=2))
    
x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
data_list = []
for data in [x0list, x1list, z0list, z1list]:
    #print(data)
    data_list.append(replace_outliers_with_neighbors(data, threshold, border=1))

for i in range(6):
    x0list, x1list, z0list, z1list = data_list[0], data_list[1], data_list[2], data_list[3]
    data_list = []
    for data in [x0list, x1list, z0list, z1list]:
        #print(data)
        data_list.append(replace_outliers_with_neighbors(data, threshold, border=0))



X,Y = np.meshgrid(muarr, alparr)

ylim1 = 5.E0
ylim2 = 4.5E1
xlim1 = 1.E0
xlim2 = 1.5E1

xtitle = r'$\mu_{0}=\nu_{0}$'
ytitle = r'$\alpha_{0,0}$'
title = r'$\frac{\alpha_{1,1}}{\alpha_{0,0}}=$'+r'{}'.format(ratioalp0alp1) + r'$\frac{\gamma_{0,0}}{\alpha_{0,0}}=$'+r'{}'.format(ratioalp0gam0) + r'$\frac{\gamma_{0,1}}{\alpha_{0,0}}=$'+r'{}'.format(ratioalp0gam01) + r'$\frac{\gamma_{1,1}}{\alpha_{0,0}}=$'+r'{}'.format(ratioalp0gam1) + r'$K=$'+r'{}'.format(K) + r'$d=$'+r'{}'.format(d)

clist = carr[:13].copy()
clist.reverse()
clist = [(0, clist[0]), (0.05, clist[1]), (0.1, clist[2]), (0.2, clist[3]), (0.32, clist[4]), (0.4, clist[5]), (0.5, clist[6]), (0.6, clist[7]), (0.65, clist[8]), (0.7, clist[9]), (0.75, clist[10]), (0.82, clist[11]), (1., clist[12])] #[(i*1/11., clist[i]) for i in range(len(clist))]
cmapphase = LinearSegmentedColormap.from_list(cmap_name, clist)
#norm = mpl.colors.Normalize(vmin=0., vmax=1.)
# Create a single figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid of subplots
fig.suptitle(title, fontsize=16)  # Main title for the figure

# List of data and labels
data_list = [(data_list[0], r'$x_0$'), (data_list[1], r'$x_1$'), (data_list[2], r'$z_0$'), (data_list[3], r'$z_1$')]
axes = axs.ravel()  # Flatten the axes array for easy iteration

# Loop through each subplot and plot
for ax, (data, label) in zip(axes, data_list):
    c = ax.pcolormesh(X, Y, data, cmap=cmapphase, vmin=0., vmax=1., shading='gouraud')
    cbar = fig.colorbar(c, ax=ax, label=label)
    cbar.set_label(label, fontsize=15)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(xlim1, xlim2)
    ax.set_ylim(ylim1, ylim2)
    ax.set_xlabel(xtitle, fontsize=15)
    ax.set_ylabel(ytitle, fontsize=15)
    ax.set_title(label, fontsize=16)
    
fig.set_size_inches(8., 8.)

# Adjust layout to prevent overlapping
fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # Leave space for the title

plt.savefig('results/phasediag_strongdilu_mualp_'+cat+str(r)+'.pdf')
```
