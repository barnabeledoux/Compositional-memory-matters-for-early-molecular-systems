```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
import scipy
from scipy.optimize import fsolve
import os
from tqdm import tqdm
import sys
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap
from class_comparts import *
from colorsarr import colorsarr
colarr = colorsarr()
carr = colarr.carr
cd = os.getcwd()

def smoothen2D(arr, neigh = 1, repet = 2):
    cop = arr.copy()
    for _ in range(repet):
        lin, col = np.size(arr, 0), np.size(arr, 1)
        for i in range(lin):
            for j in range(col):
                area = arr[max(0,i-neigh):min(lin,i+neigh), max(0,j-neigh):min(col,j+neigh)]
                cop[i,j] = np.sum(area)/area.size
        arr = cop
    return cop
```

```python
## Model A ##

Nsteps = 1
K = 1.5E1
d = 5

model = '_'

alpha = np.array([[5.E-2,0.],[0.,3.E-1]])
gamma = np.array([[1.E0,0.],[0.,2.5E0]])
mu = np.array([1.E-3,0.])
nu = np.array([1.E-3,0.])
T=2.5E0/(alpha[0][0])
slist = np.linspace(0.2, 1, 3)
siz = np.size(np.load('data/probcond_s'+str(slist[0])[:3]+'_'+model+'.npy'),0)
X0 = np.array([[[(v*u/(siz**2), v*(1-u/siz)/siz, (1-v/siz)*w/siz, (1-v/siz)*(1-w/siz)) for u in range(siz)] for v in range(siz)] for w in range(siz)])
Xbef = np.array([[[v/siz for u in range(siz)] for v in range(siz)] for w in range(siz)])
    
for s in slist:
    Prob = np.load('data/probcond_s'+str(s)[:3]+'_'+model+'.npy')

    fig = plt.figure() 
    ax = fig.add_subplot(111)

    clist = (carr[:7]+ carr[8:-2]).copy() #carr[:13].copy()
    clist.reverse()
    clist = [(0, clist[0]), (0.05, clist[1]), (0.1, clist[2]), (0.2, clist[3]), (0.3, clist[4]), (0.4, clist[5]), (0.5, clist[6]), (0.6, clist[7]), (0.7, clist[8]), (0.8, clist[9]), (1., clist[10])] #[(0, clist[0]), (0.18, clist[1]), (0.25, clist[2]), (0.31, clist[3]), (0.38, clist[4]), (0.48, clist[5]), (0.5, clist[6]), (0.6, clist[7]), (0.65, clist[8]), (0.7, clist[9]), (0.75, clist[10]), (0.82, clist[11]), (1., clist[12])]
    cmapphase = LinearSegmentedColormap.from_list('cmap', clist)

    label = 'log probability ($s={}$)'.format(str(s)[:4])
    xtitle = r'$x_0+x_1$'
    ytitle = r'$\hat{x}_0+\hat{x}_1$'
    Xabs = np.linspace(0,1,np.size(Prob,0))
    Xord = np.linspace(0,1,np.size(Prob,0))
    Yarr = np.log(smoothen2D(Prob+1E-15, neigh=2, repet=3))
    c = ax.pcolormesh(Xabs, Xord, Yarr, cmap=cmapphase, vmin=-10., vmax=0., shading='gouraud')
    cbar = fig.colorbar(c, ax=ax, label=label)
    cbar.set_label(label, fontsize=15)
    ax.set_xlabel(xtitle, fontsize=15)
    ax.set_ylabel(ytitle, fontsize=15)


    Probam = np.load('data/probcondaftmat_s'+str(s)[:3]+'_'+model+'.npy')

    fig2 = plt.figure() 
    ax2 = fig2.add_subplot(111)

    label = 'log probability ($s={}$)'.format(str(s)[:4])
    xtitle = r'$\overline{x}_0+\overline{x}_1$'
    ytitle = r'$\hat{x}_0+\hat{x}_1$'
    Xabs = np.linspace(0,1,np.size(Probam,0))
    Xord = np.linspace(0,1,np.size(Probam,0))
    Yarr = np.log(smoothen2D(Probam+1E-15, neigh=3))
    c = ax2.pcolormesh(Xabs, Xord, Yarr, cmap=cmapphase, vmin=-10., vmax=0., shading='gouraud')
    cbar = fig2.colorbar(c, ax=ax2, label=label)
    cbar.set_label(label, fontsize=15)
    ax2.set_xlabel(xtitle, fontsize=15)
    ax2.set_ylabel(ytitle, fontsize=15)
```
